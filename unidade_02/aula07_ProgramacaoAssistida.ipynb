# Soma lista

```python
# Initialize a list named 'numeros' with integer values.
numeros = [1,3]

# Initialize a variable 'soma' to store the sum, starting at 0.
soma = 0

# Iterate through each number in the 'numeros' list.
for num in numeros:
  # Add the current number to the 'soma' variable.
  soma += num

# Print the final calculated sum to the console.
print(soma)
```

  # Número primo

```python
def eh_primo(n):
  # Caso base: números menores ou iguais a 1 não são primos por definição.
  if n <= 1:
    return False
  # Casos base: 2 e 3 são números primos.
  if n <= 3:
    return True
  # Se n for divisível por 2 ou 3, não é primo (a menos que seja o próprio 2 ou 3, já tratados acima).
  # Isso elimina rapidamente múltiplos de 2 e 3.
  if n % 2 == 0 or n % 3 == 0:
    return False

  # Inicia a verificação de fatores primos a partir de 5.
  # Todos os números primos maiores que 3 podem ser expressos na forma 6k ± 1.
  i = 5
  # O loop continua enquanto i*i for menor ou igual a n.
  # Isso otimiza a verificação, pois um divisor maior que a raiz quadrada de n teria
  # um divisor correspondente menor que a raiz quadrada de n, que já teria sido encontrado.
  while i * i <= n:
    # Verifica se n é divisível por i ou por i + 2.
    # Isso cobre os possíveis fatores primos que seguem o padrão 6k ± 1.
    if n % i == 0 or n % (i + 2) == 0:
      return False
    # Incrementa i em 6 para verificar os próximos candidatos a fatores primos
    # (i.e., os números na forma 6k+1 e 6k+5).
    i += 6
  # Se nenhum divisor for encontrado após todas as verificações, o número é primo.
  return True

# Solicita ao usuário que digite um número para verificar se é primo.
numero_teste = int(input("Digite um numero para verificar se é primo: "))
# Chama a função eh_primo para determinar se o número digitado é primo.
if eh_primo(numero_teste):
  # Imprime uma mensagem informando que o número é primo.
  print(f"{numero_teste} é um número primo.")
else:
  # Imprime uma mensagem informando que o número não é primo.
  print(f"{numero_teste} não é um número primo.")
```

# Inversão de string

```python
def inverte_texto(texto):
  # A forma mais concisa e Pythonica de inverter uma string é usando slicing [::-1]
  # O slicing [::-1] cria uma cópia invertida da string original.
  # O primeiro ':' indica o início da string, o segundo ':' indica o fim,
  # e '-1' indica um passo de um caractere por vez, mas na direção inversa.
  return texto[::-1]

# Exemplo de uso da função refatorada
texto_original = "arara"
texto_invertido = inverte_texto(texto_original)
# Imprime o texto original para comparação.
print(f"O texto original é: {texto_original}")
# Imprime o texto que foi invertido pela função.
print(f"O texto invertido é: {texto_invertido}")
```

# Mini projeto

```python
def calculadora():
  # Início do bloco try-except para lidar com possíveis erros de entrada do usuário.
  try:
    # Solicita ao usuário o primeiro número e tenta convertê-lo para float.
    num01 = float(input("Digite o primeiro numero: "))
    # Solicita ao usuário o símbolo da operação desejada.
    operacao = input("Digite o símbolo da operação desejada (+, -, x, /): ")
    # Solicita ao usuário o segundo número e tenta convertê-lo para float.
    num02 = float(input("Digite o segundo numero: "))
  # Se a conversão para float falhar (por exemplo, se o usuário digitar texto),
  # uma exceção ValueError é capturada.
  except ValueError:
    # Imprime uma mensagem de erro para o usuário e encerra a função.
    print("Entrada inválida. Por favor, digite números válidos.")
    return

  # Inicializa a variável 'result' como None. Ela será usada para armazenar o resultado da operação.
  # O valor None indica que nenhuma operação foi realizada ainda.
  result = None

  # Verifica qual operação o usuário escolheu e realiza o cálculo correspondente.
  if operacao == "+":
    result = num01 + num02
  elif operacao == "-":
    result = num01 - num02
  elif operacao == "x":
    result = num01 * num02
  elif operacao == "/":
    # Verifica se o segundo número é zero antes de realizar a divisão para evitar erro.
    if num02 == 0:
      print("Erro: Divisão por zero não é permitida.")
      return # Sai da função se houver divisão por zero.
    result = num01 / num02
  else:
    # Se a operação inserida não for uma das válidas, imprime um erro.
    print("Operação inválida. Por favor, use +, -, x, ou /.")
    return # Sai da função se a operação for inválida.

  # Após todas as verificações e cálculos, imprime o resultado se uma operação válida foi realizada.
  if result is not None:
    print(f"O resultado da operação é: {result}")

# Chama a função calculadora para executá-la quando o script for rodado.
calculadora()
```
