```python
def check_temperature(threshold=30):
    # Define a função 'check_temperature' que recebe um argumento opcional 'threshold' (limite),
    # com um valor padrão de 30 graus Celsius.

    # Solicita ao usuário que insira a temperatura atual do sistema.
    while True:
        try:
            current_temp_input = input("Por favor, digite a temperatura atual do sistema em °C: ")
            current_temp = float(current_temp_input)
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número para a temperatura.")

    # Imprime a temperatura atual do sistema, formatada para duas casas decimais.
    print("Temperatura atual do sistema: " + "{:.2f}".format(current_temp) + "°C")

    # Verifica se a temperatura atual é maior que o limite definido.
    if current_temp >= threshold:
        # Se a temperatura exceder o limite, exibe um alerta de superaquecimento.
        print("ALERTA: Superaquecimento detectado! Ação necessária.")
    else:
        # Caso contrário, informa que a temperatura do sistema está normal.
        print("A temperatura do sistema está normal.")

# Chama a função 'check_temperature' para iniciar a verificação.
# Você pode ajustar o limite aqui, por exemplo, check_temperature(threshold=25).
check_temperature(threshold=30)
```

# Automação de arquivo

criat um programa que organize arquivos bagunçados em pastas automaticamente, baseado-se apenas na extensão do arquivo(ex: .pdf, .jpg)

Agora, vamos criar o programa para organizar os arquivos. Para testar, você pode criar alguns arquivos de exemplo no mesmo diretório onde o notebook está (`/content/`).

```python
import os
import shutil

def create_example_files(source_directory, extensions, num_files_per_type=10):
    """
    Cria arquivos de exemplo para teste em um diretório especificado.

    Args:
        source_directory (str): O diretório onde os arquivos de exemplo serão criados.
        extensions (list): Uma lista de extensões de arquivo (ex: ['.pdf', '.jpg']).
        num_files_per_type (int): O número de arquivos a serem criados para cada extensão.
    """
    os.makedirs(source_directory, exist_ok=True)
    print("Criando " + str(num_files_per_type) + " arquivos de exemplo para cada tipo em '" + source_directory + "'...")
    for ext in extensions:
        for i in range(1, num_files_per_type + 1):
            filename = "exemplo_" + ext[1:] + "_" + str(i) + ext
            file_path = os.path.join(source_directory, filename)
            try:
                with open(file_path, 'w') as f:
                    f.write("Conteúdo do arquivo " + filename)
                # print("Criado: " + filename) # Opcional: imprimir cada arquivo criado
            except Exception as e:
                print("Erro ao criar '" + filename + "': " + str(e))
    print("Criação de arquivos de exemplo concluída!")

def organize_files_single_block(source_directory='.', destination_base='minhas_pastas_organizadas'):
    """
    Organiza arquivos de um diretório de origem em pastas baseadas em suas extensões.
    Cria a estrutura de pastas, cria arquivos de exemplo e move os arquivos no mesmo bloco de código.

    Args:
        source_directory (str): O diretório onde os arquivos bagunçados estão e onde os exemplos serão criados.
        destination_base (str): O diretório base onde as pastas organizadas serão criadas.
    """

    print("Iniciando organização de arquivos em '" + source_directory + "' para '" + destination_base + "'...")

    # Garante que o diretório base de destino exista
    os.makedirs(destination_base, exist_ok=True)

    # Lista de extensões para as pastas que serão criadas
    # e também usada para identificar tipos de arquivos
    extensions_to_create = ['.pdf', '.jpg', '.png', '.doc', '.docx', '.xls', '.xlsx', '.txt', '.zip', '.mp3']

    # Cria arquivos de exemplo ANTES de organizar
    create_example_files(source_directory, extensions_to_create)

    # Cria uma pasta para cada tipo de extensão, se ainda não existirem
    for ext in extensions_to_create:
        folder_name = ext[1:] + '_files'
        folder_path = os.path.join(destination_base, folder_name)
        os.makedirs(folder_path, exist_ok=True)
        # print("Pasta '" + folder_name + "' criada em '" + destination_base + "'.") # Opcional: imprimir criação de cada pasta

    # Certifica-se de que a pasta 'outros_files' também exista para arquivos sem extensão conhecida
    os.makedirs(os.path.join(destination_base, 'outros_files'), exist_ok=True)

    # Lista todos os itens no diretório de origem
    for filename in os.listdir(source_directory):
        # Ignora diretórios e outros itens não relacionados a arquivos a serem organizados
        if os.path.isdir(os.path.join(source_directory, filename)):
            continue
        # if filename == os.path.basename(__file__): # Ignora o próprio script (se houver)
        #     continue # Esta linha causa o NameError em ambientes de notebook
        if filename == '.ipynb_checkpoints': # Ignora pastas de checkpoint do Jupyter/Colab
            continue
        if filename == destination_base: # Ignora a pasta de destino se estiver no mesmo nível
            continue

        # Obtém o nome completo do arquivo
        file_path = os.path.join(source_directory, filename)

        # Extrai a extensão do arquivo
        _, extension = os.path.splitext(filename)

        # Define o nome da pasta de destino com base na extensão
        if not extension or extension.lower() not in extensions_to_create:
            target_folder_name = 'outros_files'
        else:
            target_folder_name = extension[1:] + '_files'

        # Constrói o caminho completo da pasta de destino
        target_folder_path = os.path.join(destination_base, target_folder_name)

        # Constrói o caminho completo para onde o arquivo será movido
        destination_file_path = os.path.join(target_folder_path, filename)

        try:
            # Move o arquivo para a pasta de destino
            shutil.move(file_path, destination_file_path)
            print("Movido '" + filename + "' para '" + target_folder_path + "'")
        except shutil.Error as e:
            print("Erro ao mover '" + filename + "': " + str(e))
        except Exception as e:
            print("Ocorreu um erro inesperado ao mover '" + filename + "': " + str(e))

    print("Organização de arquivos concluída!")

# Para executar a função de organização:
# Certifique-se de que o diretório onde você quer organizar os arquivos
# (por padrão, o diretório atual '.') contém os arquivos que você quer mover.
# A função irá mover esses arquivos para as subpastas dentro de 'minhas_pastas_organizadas'.
organize_files_single_block()
```

# Consulta simples a APIs

desenvolver um script que consulte uma API publica (ViaCEP) e retorne as informações de endereços formatadas de maneira legivel

```python
import requests

def consultar_cep_brasilapi():
    """
    Consulta a BrasilAPI para obter informações de endereço a partir de um CEP fornecido pelo usuário.
    A BrasilAPI agrega dados de diversas fontes, incluindo os Correios.
    Trata erros de conexão e CEPs inválidos.
    """
    while True:
        cep = input("Digite o CEP (apenas números, 8 dígitos) ou 'sair' para encerrar: ").strip()

        if cep.lower() == 'sair':
            print("Consulta encerrada.")
            return

        # Validação básica do CEP
        if not cep.isdigit() or len(cep) != 8:
            print("CEP inválido. Por favor, digite 8 dígitos numéricos.")
            continue

        # Usando a BrasilAPI
        url = f"https://brasilapi.com.br/api/cep/v1/{cep}"

        try:
            response = requests.get(url, timeout=5) # Adiciona um timeout para evitar que a requisição trave
            response.raise_for_status() # Lança uma exceção para códigos de status HTTP de erro (4xx ou 5xx)

            data = response.json()

            # BrasilAPI retorna 404 para CEPs não encontrados, capturado por raise_for_status()
            # Caso contrário, o CEP foi encontrado.
            print("\n--- Detalhes do Endereço (BrasilAPI) ---")
            print(f"CEP: {data.get('cep', 'N/A')}")
            print(f"Logradouro: {data.get('street', 'N/A')}")
            print(f"Bairro: {data.get('neighborhood', 'N/A')}")
            print(f"Cidade: {data.get('city', 'N/A')}")
            print(f"Estado (UF): {data.get('state', 'N/A')}")
            print(f"Serviço de Origem: {data.get('service', 'N/A')}") # Indica a fonte do dado (ex: correios)
            print("---------------------------\n")

        except requests.exceptions.Timeout:
            print("Erro: A requisição excedeu o tempo limite. Verifique sua conexão ou tente novamente mais tarde.")
        except requests.exceptions.ConnectionError:
            print("Erro: Falha na conexão com a internet. Verifique sua conexão.")
        except requests.exceptions.RequestException as e:
            if response.status_code == 404:
                print(f"Erro: CEP '{cep}' não encontrado ou inválido pela BrasilAPI.")
            else:
                print(f"Erro na requisição HTTP: {e}")
        except ValueError:
            print("Erro ao decodificar a resposta JSON. A API pode ter retornado um formato inesperado.")
        except Exception as e:
            print(f"Ocorreu um erro inesperado: {e}")

# Chamando a nova função para iniciar a consulta
consultar_cep_brasilapi()
```

# Sistema de notificação

criar um programa autônomo que dispare alertas(mensagens de alerta no console ou e-mail simulado)unicamente quando uma condição crítica for atingida

### Implementação do Sistema de Notificação Autônomo

Vamos criar um programa que simula o monitoramento de um sistema e dispara alertas quando uma condição crítica é detectada. Por enquanto, os alertas serão exibidos no console, mas a estrutura pode ser adaptada para enviar e-mails ou outras notificações.

```python
import time # Importa o módulo time para usar a função sleep (pausar a execução)
import random # Importa o módulo random para gerar números aleatórios

def check_critical_condition(threshold=30): # Define a função que verifica a condição crítica, com um valor padrão de 30 graus Celsius.
    """
    Simula a leitura de uma condição climática (ex: temperatura ambiente) e verifica se uma condição crítica foi atingida.
    Retorna True se a condição for crítica, False caso contrário.
    """
    # Simula uma leitura de temperatura ambiente aleatória entre 15 e 40 graus Celsius
    current_reading = random.uniform(15, 40) # Gera um número float aleatório
    print("Leitura atual da temperatura ambiente: " + "{:.2f}".format(current_reading) + "°C") # Exibe a temperatura atual formatada para duas casas decimais.

    # Compara a leitura atual com o limiar crítico
    if current_reading > threshold:
        return True # Retorna True se a temperatura exceder o limite
    else:
        return False # Retorna False caso contrário

def send_console_alert(message):
    """
    Simula o envio de um alerta para o console.
    """
    print("!!! ALERTA CLIMÁTICO: " + message + " !!!") # Imprime a mensagem de alerta no console


def monitor_system(interval_seconds=5, critical_threshold=35, max_alerts=3):
    """
    Monitora o sistema continuamente e dispara alertas quando uma condição crítica é atingida.

    Args:
        interval_seconds (int): Intervalo de tempo entre as verificações em segundos.
        critical_threshold (int): O valor acima do qual a condição é considerada crítica (temperatura).
        max_alerts (int): O número máximo de alertas a serem disparados antes de parar o monitoramento.
    """
    print("Iniciando monitoramento de eventos climáticos a cada " + str(interval_seconds) + " segundos...") # Informa o início do monitoramento
    print("Condição crítica de temperatura acima de " + str(critical_threshold) + "°C.") # Informa o limiar crítico

    alerts_sent = 0 # Inicializa o contador de alertas enviados
    while alerts_sent < max_alerts: # Loop continua enquanto o número de alertas enviados for menor que o máximo permitido
        # Verifica se a condição crítica foi atingida
        if check_critical_condition(threshold=critical_threshold):
            send_console_alert("Temperatura ambiente excedeu o limite crítico!") # Envia o alerta se for crítico
            alerts_sent += 1 # Incrementa o contador de alertas
        time.sleep(interval_seconds) # Pausa a execução pelo tempo definido antes da próxima verificação

    print("Monitoramento encerrado após disparar " + str(alerts_sent) + " alertas climáticos.") # Informa o fim do monitoramento

# --- EXECUTAR MONITORAMENTO ---
# Chamada da função para iniciar o monitoramento do sistema com os parâmetros definidos:
# - interval_seconds=2: verifica a condição a cada 2 segundos.
# - critical_threshold=35: dispara alerta se a leitura de temperatura for acima de 35°C.
# - max_alerts=2: para o monitoramento após 2 alertas serem disparados.
monitor_system(interval_seconds=2, critical_threshold=35, max_alerts=2) # Inicia o monitoramento com os parâmetros especificados
```

### Como usar o sistema de notificação

A função `monitor_system` simula um processo de monitoramento. Ela verifica uma condição a cada `interval_seconds` e dispara um alerta se `check_critical_condition` retornar `True`. O monitoramento é interrompido após `max_alerts` serem enviados.

Para executar, basta descomentar a última linha do código acima e rodar a célula. Você pode ajustar os parâmetros `interval_seconds`, `critical_threshold` e `max_alerts` conforme necessário.
